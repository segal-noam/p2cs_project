# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

# Desired Heatmap

```python
# Generate random 9x9 data with values between 0 and 1
np.random.seed(1)
data = np.random.rand(9, 9)

# Set main diagonal to 1 (to show up red on 'coolwarm' with vmin=0, vmax=1)
for d in range(9):
    data[d, d] = 1

fig, ax = plt.subplots(figsize=(8,8))

# Show heatmap
im = ax.imshow(data, cmap='coolwarm', vmin=0, vmax=1)

# Remove ticks
ax.set_xticks([])
ax.set_yticks([])

# Label axes
ax.set_xlabel(r'$\bf{HKs}$', fontsize=20, labelpad=15)
ax.set_ylabel(r'$\bf{RRs}$', fontsize=20, labelpad=15)

# Add colorbar
cbar = plt.colorbar(im, ax=ax, shrink=0.8, pad=0.03)

# Add colored line segments next to row indices (left)
for i in range(9):
    color = 'blue' if i < 4 else 'orange'
    # Rectangles are now wider and on top for visibility, linewidth enhanced if needed
    rect = patches.FancyBboxPatch(
        (-1.18, i-0.42),   # further left for guaranteed visibility, align with heatmap cell
        0.32, 0.84,        # wider and slightly taller
        boxstyle="round,pad=0.03",
        linewidth=1,
        edgecolor='none',
        facecolor=color,
        zorder=3,
        mutation_aspect=0.5
    )
    ax.add_patch(rect)

# Add colored line segments next to column indices (top)
for j in range(9):
    color = 'blue' if j < 4 else 'orange'
    rect = patches.FancyBboxPatch(
        (j-0.42, -1.18),   # higher up for guaranteed visibility
        0.84, 0.32,        # wider and slightly taller
        boxstyle="round,pad=0.03",
        linewidth=1,
        edgecolor='none',
        facecolor=color,
        zorder=3,
        mutation_aspect=0.5
    )
    ax.add_patch(rect)

plt.tight_layout()

plt.savefig("ideal_rr_hk_heatmap.png", dpi=300)

plt.show()
```
